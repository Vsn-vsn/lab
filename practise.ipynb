```python
import random as r
class NQHillClimb:
    def __init__(self,N):
        self.N=N

    def printing(self,state):
        for r in range(self.N):
            row=""
            for col in range(self.N):
                row+="1 " if state[col]==r else "0 "
            print(row)
            print()

    def getneighbours(self,state):
        neighbours=[]
        for j in range(self.N):
            for i in range(self.N):
                if i!=state[j]:
                    new_state=state.copy()
                    new_state[j]=i
                    neighbours.append(new_state)
        return neighbours
    
    def getheuristic(self,state):
        attacks=0

        for i in range(self.N):
            for j in range(i+1,self.N):
                attacks+=1 if abs(state[i]-state[j])==abs(i-j) or state[i]==state[j] else 0
        return attacks
    
    def hill_climb_solve_N_queens(self):
        resets=0
        while True:
            cur_state=[r.randint(0,self.N-1) for _ in range(self.N)]
            cur_h=self.getheuristic(cur_state)
            
            while True:
                neighbour=self.getneighbours(cur_state)
                neighbour_h=[(self.getheuristic(n),n) for n in neighbour]
                neighbour_h.sort()

                best_h, best_state=neighbour_h.pop(0)
                if best_h >= cur_h:
                    break
                cur_h=best_h
                cur_state=best_state

            if cur_h==0:
                print(f"Restarts: {resets}")
                return cur_state
            resets+=1
            
hl=NQHillClimb(8)
sol=hl.hill_climb_solve_N_queens()
print("Solution with heuristic= ",hl.getheuristic(sol))
hl.printing(sol)
```

```python
# ucs
class UCSGraph:
    def __init__(self):
        self.graph={}
    
    def add_edge(self,u,v,w):
        if u not in self.graph:
            self.graph[u]=[]
        self.graph[u].append((v,w))
    
    def ucs_search(self,start,goal):
        visited=set()
        pq=[(0,start,[])]

        while pq:
            pq.sort()
            weight,node,path=pq.pop(0)

            path=path+[node]

            if node in visited:
                continue
            visited.add(node)

            if node==goal:
                return path,weight
            
            for n,w in self.graph.get(node,[]):
                if n not in visited:
                    pq.append((w+weight,n,path))
        return None,float("inf")

ucs=UCSGraph()
ucs.add_edge("Arad","Sibiu",140)
ucs.add_edge("Arad","Zerind",75)
ucs.add_edge("Arad","Timisoara",118)
ucs.add_edge("Zerind","Oradea",71)
ucs.add_edge("Timisoara","Lugoj",111)
ucs.add_edge("Lugoj","Mehadia",70)
ucs.add_edge("Mehadia","Dobreta",75)
ucs.add_edge("Dobreta","Craiova",120)
ucs.add_edge("Craiova","Pitesti",138)
ucs.add_edge("Pitesti","Bucharest",101)
ucs.add_edge("Bucharest","Giurgiu",90)
ucs.add_edge("Sibiu","Rimni Vilcea",80)
ucs.add_edge("Rimni Vilcea","Craiova",146)
ucs.add_edge("Rimni Vilcea","Pitesti",97)
ucs.add_edge("Sibiu","Fagaras",99)
ucs.add_edge("Sibiu","Oradea",151)
ucs.add_edge("Fagaras","Bucharest",211)

path,cost=ucs.ucs_search("Arad","Bucharest")
if path:
    print(f"Path: {path}")
    print(f"Cost:{cost}")
else:
    print("No path")
```

```python
from heapq import heappush as push, heappop as pop
class AStar:
    def __init__(self):
        self.graph={}
        self.heuristic={}
    
    def addedge(self, u, v, w):
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []  # ensure v exists in graph
        self.graph[u].append((v, w))


    def addheu(self,u,h):
        self.heuristic[u]=h

    def astar(self,start,goals):
        pq=[]
        push(pq,(0+self.heuristic[start],start,0))
        g={n: float("inf") for n in self.graph}
        g[start]=0
        paths={}

        while pq:
            _,cur_node,cur_g=pop(pq)
            if cur_node in goals:
                return self.pathConstruct(paths,cur_node) # type: ignore
            
            for n,c in self.graph[cur_node]:
                temp=cur_g+c
                if temp < g[n]:
                    f=temp+self.heuristic[n]
                    paths[n]=cur_node
                    g[n]=temp
                    push(pq,(f,n,temp))
        return None
    def pathConstruct(self,paths,cur):
        path=[cur]
        i=cur
        while i in paths:
            i=paths[i]
            path.append(i)
        return path[::-1]
    
g = AStar()
g.addedge('A','B',3)
g.addedge('A','G1',9)
g.addedge('B','A',3)
g.addedge('S','A',5)
g.addedge('S','B',9)
g.addedge('S','D',6)
g.addedge('D','S',1)
g.addedge('D','C',2)
g.addedge('C','S',6)
g.addedge('B','C',1)
g.addedge('C','G2',5)
g.addedge('C','F',7)
g.addedge('F','D',2)
g.addedge('D','E',2)
g.addedge('E','G3',7)
g.addedge('F','G3',8)

g.addheu('A',7)
g.addheu('B',3)
g.addheu('C',4)
g.addheu('D',6)
g.addheu('E',5)
g.addheu('F',6)
g.addheu('G1',0)
g.addheu('G2',0)
g.addheu('G3',0)
g.addheu('S',5)

sol=g.astar('S',['G1','G2','G3'])
if sol:
    print(sol)
else:
    print("no solution")
```

```python
class DFSGraph:
    def __init__(self,vertices):
        self.vertices=vertices
        self.graph={v:[] for v in range(vertices)}

    def addedge(self,u,v):
        self.graph[u].append(v)
        self.graph[v].append(u)

    def dfshelper(self,visited,node):
        visited.add(node)
        print(node, end="   ")
        for n in self.graph[node]:
            if n not in visited:
                self.dfshelper(visited,n)
    
    def dfs(self):
        visited=set()
        print("DFS: ",end="   ")
        for node in self.graph:
            if node not in visited:
                self.dfshelper(visited,node)

gr=DFSGraph(5)
gr.addedge(0,2)
gr.addedge(2,4)
gr.addedge(0,3)
gr.addedge(1,0)  


gr.dfs()
```

```python
#maze
class MAZE:
    def __init__(self):
        self.graph={}
        self.path=[]
    
    def addtile(self,u,v):
        if u not in self.graph:
            self.graph[u]=[]
        if v not in self.graph:
            self.graph[v]=[]
        self.graph[u].append(v)
        self.graph[v].append(u)
    
    def findsol(self,node,goal,visited):
        visited.add(node)
        self.path.append(node)
        if node==goal:
            return True
        
        for n in self.graph[node]:
            if n not in visited:
                if self.findsol(n,goal,visited):
                    return True
        
        self.path.pop()
        return False

sol=MAZE()
sol.addtile(1,6)
sol.addtile(1,2)
sol.addtile(2,3)
sol.addtile(3,8)
sol.addtile(4,5)
sol.addtile(5,10)
sol.addtile(6,11)
sol.addtile(7,8)
sol.addtile(9,14)
sol.addtile(9,10)
sol.addtile(10,15)
sol.addtile(11,12)
sol.addtile(12,17)
sol.addtile(13,14)
sol.addtile(14,19)
sol.addtile(15,20)
sol.addtile(16,17)
sol.addtile(17,18)
sol.addtile(18,19)

visited=set()
solver=sol.findsol(2,20,visited)
if solver:
    print(sol.path)
else:
    print("not solvable")
```

```python
#BFS

class BFS:
    def __init__(self,vertices):
        self.vertices=vertices
        self.graph={g:[] for g in range(vertices)}
        self.indeg={g:0 for g in range(vertices)}

    def addedge(self,u,v):
        self.graph[u].append(v)
        self.indeg[v]+=1
    
    def bfs_traversal(self,start):
        topo=[]
        queue=[start]
        
        while queue:
            u=queue.pop(0)

            print(u,end="   ")
            for n in self.graph[u]:
                queue.append(n)

    def toposort(self):
        queue = []
        topo = []

        for node in self.graph:
            if self.indeg[node] == 0:
                queue.append(node)

        while queue:
            u = queue.pop(0)
            topo.append(u)

            for v in self.graph[u]:
                self.indeg[v] -= 1
                if self.indeg[v] == 0:
                    queue.append(v)
        print("\nTopological:",topo)

gr=BFS(6)
gr.addedge(5,0)
gr.addedge(4,0)
gr.addedge(5,2)
gr.addedge(2,3)  
gr.addedge(3,1)
gr.addedge(4,1)

gr.bfs_traversal(5)
gr.toposort()
```

```python
# TSP using bfs
def bfs_traverse(start,graph):
    queue=[(start,[start],0)]
    min_cost=float("inf")
    best_path=[]

    while queue:
        city,path,cost=queue.pop(0) 

        if len(path)==len(graph):
            if start in graph:
                total_cost=cost+graph[city][start]
                if total_cost<min_cost:
                    min_cost=total_cost
                    best_path=path+[start]
            continue
        
        for n in graph[city]:
            if n not in path:
                queue.append((n,path+[n],cost+graph[city][n]))
    return best_path,total_cost
graph = { 
'A': {'B': 2, 'C': 3, 'D': 1}, 
'B': {'A': 2, 'C': 4, 'D': 2}, 
'C': {'A': 3, 'B': 4, 'D': 3}, 
'D': {'A': 1, 'B': 2, 'C': 3} 
}

pth,tot=bfs_traverse('A',graph)
print(pth)
print(tot)
```
