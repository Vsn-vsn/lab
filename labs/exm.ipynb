```python
import random
def f(x):
    return x**2 - 5*x + 6

def hill_climb(a,b, step,iterations):
    #let us take a start value of x
    x=random.uniform(a,b)
    print(f"start value of : {x}")

    for _ in range(iterations):
        new_x= x + random.uniform(-step, step)
        if a<= new_x <= b and f(new_x) > f(x):
            x=new_x
    
    return x, f(x)

x,v=hill_climb(-10,10,0.01,1000)
print(f"The max value of the function is {v} whenn x={x}")
```

```python
state_space = {
    'A': [('T', 11), ('B', 13), ('C', 21)],
    'B': [('D', 27), ('E', 3)],
    'C': [('F', 25), ('G', 4)],
    'D': [('H', 101), ('I', 99)],
    'F': [('J', 67)],
    'G': [('K', 99), ('L', 3)],
    'H': [('M', 17)],
    'I': [('M', 17)],
    'J': [('M', 17)]
}

heuristics = {
    'A': 5, 'T': 11, 'B': 13, 'C': 21, 'D': 27, 'E': 3,
    'F': 25, 'G': 4, 'H': 101, 'I': 99, 'J': 67, 'K': 99,
    'L': 3, 'M': 17
}

def get_gen(node):
    """ Returns the children of a given node """
    return state_space.get(node, [])

def sort_children(lst):
    """ Sorts children based on heuristic in descending order """
    lst.sort(key=lambda x: x[1], reverse=True)

def hillClimb(start, goal):
    closed = []
    N = (start, heuristics[start])  
    print(f"\nInitial Node: {N}")

    while True:
        closed.append(N)
        child = get_gen(N[0])
        sort_children(child)

        print(f"\nNode: {N}")
        print("Sorted Children:", child)
        print("Closed List:", closed)

        if N[0] == goal:
            print("\nGoal reached!")
            return
        
        if not child or child[0][1] <= N[1]:  # No better move
            print("\nStuck in local maximum. No more moves.")
            return

        N = child[0]  # Pick the best move

hillClimb('A', 'H')
```

```python
# A* algo
from heapq import heappop, heappush

class GraphA:
    def __init__(self):
        self.graph={}
        self.heu={}
    
    def addedge(self,u,v,cost):
        if u not in self.graph:
            self.graph[u]=[]
        if v not in self.graph:
            self.graph[v]=[]
        self.graph[u].append((v,cost))
        self.graph[v].append((u,cost))

    def heuristics(self,u,val):
        self.heu[u]=val

    def a_star(self,start,goal):
        openlist=[]
        heappush(openlist,(0+self.heu[start],start,0))
        g_score={n: float('inf') for n in self.graph}
        g_score[start]=0
        paths={}

        while openlist:
            _,cur, cur_g=heappop(openlist)

            if cur == goal:
                return self.path_construct(paths,cur)
            
            for n, c in self.graph[cur]:
                temp_g=cur_g +c
                if temp_g < g_score[n]:
                    paths[n]=cur
                    g_score[n]=temp_g
                    f=temp_g+self.heu[n]
                    heappush(openlist,(f,n,temp_g))
        return None

    def path_construct(self,paths,cur):
        pth=[cur]
        while cur in paths:
            cur=paths[cur]
            pth.append(cur)
        return pth[::-1]
    
graph = GraphA()

# Defining the graph (Node, Cost)
graph.addedge('A', 'B', 4)
graph.addedge('A', 'C', 3)
graph.addedge('B', 'D', 2)
graph.addedge('C', 'D', 5)
graph.addedge('B', 'E', 10)
graph.addedge('D', 'E', 3)

# Defining heuristic values
graph.heuristics('A', 7)
graph.heuristics('B', 6)
graph.heuristics('C', 2)
graph.heuristics('D', 1)
graph.heuristics('E', 0)  # Goal node has heuristic 0

path = graph.a_star('A', 'E')
print("Shortest Path:", path)
```

```python
def water_jug_bfs(jug1_cap, jug2_cap, target):
    queue = [(0, 0)]  # initially both jugs are empty
    visited = set([(0,0)])
    parent = {}

    while queue:
        x, y = queue.pop(0)  # dequeue operation

        if x == target :
            path = []
            while (x, y) in parent:
                path.append((x, y))
                x, y = parent[(x, y)]
            path.append((0, 0))
            return path[::-1]

        states = [
            (jug1_cap, y),  # Fill jug1
            (x, jug2_cap),  # Fill jug2
            (0, y),         # Empty jug1
            (x, 0),         # Empty jug2
            (x - min(x, jug2_cap - y), y + min(x, jug2_cap - y)),   # Pour jug1 -> jug2
            (x + min(y, jug1_cap - x), y - min(y, jug1_cap - x))    # Pour jug2 -> jug1
        ]

        for state in states:
            if state not in visited:
                queue.append(state)
                visited.add(state)
                parent[state] = (x, y)
    
    return None



j1=4
j2=3
t=2
pt=water_jug_bfs(j1,j2,t)
if pt:
    print(pt)
else:
    print("No solution")
```

```python
def water_jug_dfs(j1,j2,target):
    stack=[(0,0)]
    visited=set([(0,0)])
    parent={}

    while stack:
        x,y = stack.pop()

        if x==target:
            path=[]
            while (x,y) in parent:
                path.append((x,y))
                x,y=parent[(x,y)]
            path.append((0,0))
            return path[::-1]
        
        states=[
            (j1,y),
            (x,j2),
            (x,0),
            (0,y),
            (x-min(x,j2-y), y+min(x,j2-y)),
            (x+min(y,j1-x),y-min(y,j1-x))
        ]

        for s in states:
            if s not in visited:
                stack.append(s)
                visited.add(s)
                parent[s]=(x,y)
    return None

j1=4
j2=3
t=2
pt=water_jug_dfs(j1,j2,t)
if pt:
    print(pt)
else:
    print("No solution")
```

```python
# N-queens using hill climning technique
import random
N=8

def print_board(state):
    for r in range(N):
        row=""
        for c in range(N):
            row+="1 " if state[c]==r else "0 "
        print(row)
    print()

def heuri(state):
    attacks=0
    for i in range(N):
        for j in range(i+1, N):
            attacks+=1 if state[i]==state[j] or abs(state[i]-state[j])==abs(i-j) else 0
    return attacks

def getNeighbours(state):
    neighbours=[]
    for j in range(N):
        for i in range(N):
            if i!=state[j]:
                ns=state.copy()
                ns[j]=i
                neighbours.append(ns)
    return neighbours

def hillQueenClimb():
    rest=0
    while True:
        rest+=1
        cur=[random.randint(0,N-1) for _ in range(N)]
        cur_h=heuri(cur)

        while True:
            neighbours=getNeighbours(cur)
            nbr_h=[(heuri(n),n) for n in neighbours]
            nbr_h.sort()

            best_h,bes_n=nbr_h[0]
            if best_h >=cur_h:
                #return cur
                break
            
            cur_h=best_h
            cur=bes_n
        if cur_h == 0:
            print(f"Solved after {rest} random restarts.")
            return cur


solution = hillQueenClimb()
print("Final board with heuristic =", heuri(solution))
print_board(solution)
        
```
